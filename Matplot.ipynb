```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
prumerna_tf = np.mean(tf_test, axis=0)

plt.figure(figsize=(10, 5))

plt.plot(prumerna_tf, color='blue')
plt.title("Průměrná tepová frekvence pacientů během zátěžového testu")
plt.xlabel("Minuty")
plt.ylabel("Tepová frekvence (BPM)")
plt.show()
```

```python
faze_data = [
    tf_test[:, 0:10].ravel(),
    tf_test[:, 10:20].ravel(),
    tf_test[:, 20:40].ravel(),
    tf_test[:, 40:50].ravel(),
    tf_test[:, 50:60].ravel()
]
faze_labels = ["Klid", "Zahřívání", "Aerobní", "Sprint", "Regenerace"]

plt.figure(figsize=(10, 5))

plt.boxplot(faze_data, labels=faze_labels)
plt.title("Rozložení tepové frekvence v jednotlivých fázích testu")
plt.ylabel("Tepová frekvence (BPM)")
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.hist(faze_data[0], bins=15, alpha=0.7, label="Klid", color='green')
plt.hist(faze_data[3], bins=15, alpha=0.7, label="Sprint", color='red')

plt.title("Porovnání četností tepových frekvencí: Klid vs. Sprint")
plt.xlabel("Tepová frekvence (BPM)")
plt.ylabel("Počet záznamů")
plt.legend()
plt.show()
```
